**Understanding Every Feature in Data**

Memahami makna data. Berikut merupakan fitur dari data:

*   PassengerId = Id Penumpang

*   pclass = Ticket class 1=1st, 2=2nd, 3=3rd
*   Name = Nama dari setiap penumpang
*   Sex = Jenis Kelamin Penumpang
*   Age = Umur dari Penumpang
*   Fare = Tarif
*   Sibsp of siblings / spouses = Jumlah saudara atau Pasangan
*   Parch = of parents / children = Jumlah Orang tua atau anak
*   Ticket = Ticket number
*   Cabin = Cabin number = Nomor Kamar di kapal
*   Embarked = Port of Embarkation / Pelabuhan Pemberangkatan (C=Cherbourg, Q=Queenstown, S=Southampton)

```python
def group_barchart(column):
  survived = train.groupby([column, "Survived"]).count()["PassengerId"].reset_index()

  g = sns.catplot(
      data= survived, kind = "bar",
      x = "Survived", y = "PassengerId", hue= column,
      palette="dark", height=6
  )
```

**Mengisi missing value categorical data**

1. Mengisi missing value embarked (data train)

```python
train[train["Embarked"].isnull()]
```

```python
train['Embarked'].value_counts()
```

```python
modus_embarked = train['Embarked'].mode()[0]
modus_embarked
```

mengisis data pada fitur **Embarked** dengan cara modus (nilai paling sering muncul)

```python
train['Embarked'] = train['Embarked'].fillna(modus_embarked)
```

```python
train[train["Embarked"].isnull()]
```

2. Mengisi missing value age (data train)

```python
train[train["Age"].isnull()]
```

karena data age merupakan numerik, maka untuk mengisi missing value nya menggunakan mean (rata-rata).

```python
mean_age = train['Age'].mean()
mean_age
```

```python
train['Age'] = train['Age'].fillna(mean_age)
```

```python
train[train["Age"].isnull()]
```

3. Mengisi missing value age (data test)

```python
test[test["Age"].isnull()]
```

```python
mean_age_test = test['Age'].mean()
mean_age_test
```

```python
test['Age'] = test['Age'].fillna(mean_age_test)
```

```python
test[test["Age"].isnull()]
```

4. Mengisi missing value fare (data test)

```python
test[test["Fare"].isnull()]
```

```python
mean_fare_test = test['Fare'].mean()
mean_fare_test
```

```python
test['Fare'] = test['Fare'].fillna(mean_fare_test)
```

```python
test[test["Fare"].isnull()]
```

5. Drop/Hapus fitur Cabin, karena terlalu banyak missing value

```python
train = train.drop(['Cabin'], axis=1)
test = test.drop(['Cabin'], axis=1)
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

**Convert Categorical Data To Numberical Data**

1. Convert Data to Sex (biner)

```python
train.head()
```

```python
test.head()
```

```python
sex_mapping = {"male":1, "female":0}

train['Sex'] = train['Sex'].map(sex_mapping)
test['Sex'] = test['Sex'].map(sex_mapping)
```

```python
train.head()
```

2. Convert Data Embarked train (one hot encoding)

```python
embarked = train[('Embarked')]
dummies = pd.get_dummies(embarked, prefix='Embarked', dtype=int)

dummies.head()
```

```python
train.head()
```

```python
train = pd.merge(train.reset_index(), dummies.reset_index())
train.head()
```

```python
##menghapus index dan embarked
train = train.drop(['index', 'Embarked'], axis=1)
```

3. Convert Data Embarked test (one hot encoding)

```python
embarked_test = test[('Embarked')]
dummies_test = pd.get_dummies(embarked_test, prefix='Embarked', dtype=int)

dummies_test.head()
```

```python
test = pd.merge(test.reset_index(), dummies_test.reset_index())
test.head()
```

```python
##menghapus index dan embarked
test = test.drop(['index', 'Embarked'], axis=1)
```

```python
test.head()
```

4. Extract Data Title From Name and Convert Title

```python
train['Title'] = train['Name'].str.extract('([A-Za-z]+)\.', expand=False)
```

```python
train.head()
```

Menggunakan ordinal untuk mengurutkan Mr,Mrs,Miss,...

```python
tittle_mapping =  {
    "Mr": 0,
    "Miss": 1,
    "Mrs": 2,
    "Master": 3,"Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
    "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3}


train['Title'] = train['Title'].map(tittle_mapping)
```

```python
train.head()
```

```python
test['Title'] = test['Name'].str.extract('([A-Za-z]+)\.', expand=False)
```

```python
test['Title'] = test['Title'].map(tittle_mapping)
```

## Modelling

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
train.shape, test.shape
```

```python
#menentukan target
train_data = train.drop(['Survived'], axis=1)
target = train['Survived']
```

```python
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)

clf = GaussianNB()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
score.mean()
```

```python
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)

clf = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
score.mean()
```

```python
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)

clf = RandomForestClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
score.mean()
```

```python
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)

clf = KNeighborsClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
score.mean()
```

```python
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)

clf = SVC()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
score.mean()
```
